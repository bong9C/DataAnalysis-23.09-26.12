```python
# 대한민국 인구
kor_population = kor_date_index_df['population']['2022-01-01']

# 미국 인구
usa_population = usa_date_index_df['population']['2022-01-01']

# 폴란드 인구
pol_population = pol_date_index_df['population']['2022-01-01']

# 이탈리아 인구
ita_population = ita_date_index_df['population']['2022-01-01']

# 프랑스 인구
fra_population = fra_date_index_df['population']['2022-01-01']
```

```python
# 인구 비율 구하기
rate1 = round((kor_population/usa_population), 2)
rate2 = round((kor_population/pol_population), 2)
rate3 = round((kor_population/fra_population), 2)
rate4 = round((kor_population/ita_population), 2)
```

```python

final_df = pd.DataFrame(
   {
        "KOR": kor_total_cases,
        "FRA": fra_total_cases * rate3,
        "USA" :usa_total_cases * rate1,
        "POL" :pol_total_cases * rate2,
        "ITA" :ita_total_cases * rate4,
     }, index=kor_date_index_df.index)

# 일부 데이터 확인

final_df.head()
```
