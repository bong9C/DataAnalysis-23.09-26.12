```python
# 원하는 열 리스트 만들기
selected_columns = ['iso_code', 'location', 'date','total_cases','population']

# 원하는 열로 개정된 데이터프레임 생성
revised_df = df[selected_columns]

# 개정된 데이터프레임 데이터 일부 확인하기
revised_df.head()
```

```python
# 열 location 값 확인하기
locations = revised_df['location']
locations
```

```python
# 대한민국 데이터프레임 생성하기
kor_df = revised_df[revised_df.location == 'South Korea']

# 데이터 일부 확인하기 
kor_df.head()
```

```python
# 색인 면경하기
kor_date_index_df = kor_df.set_index('date')

# 데이터 일부 확인하기
kor_date_index_df.head()
```

```python
# 미국 데이터프레임 생성하기
usa_df = revised_df[revised_df.location == 'United States']

# 색인 변경하기
usa_date_index_df = usa_df.set_index('date')

# 값 일부 확인하기
usa_date_index_df.head()
```

```python
df = pd.DataFrame(
   {
        "pig": [20, 18, 489, 675, 1776],
        "horse": [4, 25, 281, 600, 1900],
     }, index=[1990, 1997, 2003, 2009, 2014])
lines = df.plot.line()
```

```python
 # 대한민국 확진자 수 구하기

kor_total_cases = kor_date_index_df['total_cases']
kor_total_cases
```

```python
# 미국 확진자 수 구하기

usa_total_cases = usa_date_index_df['total_cases']
usa_total_cases
```

```python
# 색인 확인하기
kor_date_index_df.index
```

```python
# 시각화용 데이터프레임 생성

final_df = pd.DataFrame(
   {
        "KOR": kor_total_cases,
        "USA": usa_total_cases,
     }, index=kor_date_index_df.index)

# 일부 데이터 확인

final_df.head()
```

```python
# 선 그래프 그리기
final_df.plot.line()
```

```python
# 올해 그래프 그리기

final_df['2022-01-01':].plot.line()
```

```python
# x축 라벨 각도 변경하기 
final_df['2022-01-01':].plot.line(rot=45)
```

```python
# 대한민국 인구
kor_population = kor_date_index_df['population']['2022-01-01']

# 미국 인구
usa_population = usa_date_index_df['population']['2022-01-01']

print('미국 인구(', usa_population, ') /  대한민국 인구(', kor_population, ')')
```

```python
# 인구 비율 구하기
rate = round((usa_population / kor_population), 2)
rate
```

```python
# 확진자 비율을 적용한 최종 데이터 프레임 생성기
final_revised_df = pd.DataFrame({
    'KOR': kor_total_cases * rate,
    'USA' : usa_total_cases
}, index=kor_date_index_df.index
)

# 일부 데이터 확인하기
final_revised_df.head(['date'],['2022-01-01'])
```
