```python
import requests  # HTTP 요청을 보내기 위한 라이브러리 불러오기
import json

url = 'http://api.vworld.kr/req/address'  # VWorld 주소 → 좌표 변환 API 기본 URL
primary_key = ' DAC1D267-E17C-30B7-B765-F63136205DD1'  # API 인증키

def request_geo(road, addr_type='ROAD'):  # 주소(road)와 주소유형(기본값: 도로명)을 인자로 받는 함수
    params = {
        'service': 'address',    # 사용할 서비스: 주소 서비스
        'request': 'getcoord',   # 요청 유형: 좌표 가져오기
        'version': '2.0',        # API 버전
        'crs': 'epsg:4326',      # 좌표계: WGS84 (위경도 방식)
        'refine': 'true',        # 주소 정제 여부: 사용 -불완전하거나 약식이어도 자동 보정
        'simple': 'false',       # 간략 응답 여부: 상세 결과 반환
        'format': 'json',        # 응답 데이터 형식: JSON
        'type': addr_type,       # 주소 유형: ROAD(도로명) 또는 PARCEL(지번)
        'address': road,         # 좌표로 변환할 주소값
        'key': primary_key       # API 인증키
    }

    page = requests.get(url, params=params)  # API URL에 파라미터를 붙여 GET 요청 전송
    json_data = page.json()  # 응답 결과를 JSON(딕셔너리) 형태로 파싱
    print(json.dumps(json_data, ensure_ascii=False, indent=2)) #한글 깨짐 방지, 들여쓰기 2칸

    if json_data['response']['status'] == 'OK':  # 응답 상태가 정상(OK)인 경우
        x = json_data['response']['result']['point']['x']  # 경도(longitude) 값 추출
        y = json_data['response']['result']['point']['y']  # 위도(latitude) 값 추출
        return x, y  # 경도, 위도 반환
    else:
        return 0, 0  # 요청 실패 시 기본값 0, 0 반환

# 도로명 주소로 좌표 요청 후 x(경도), y(위도)에 저장
x, y = request_geo("인천광역시 부평구 무네미로448번길 56 (구산동, 한국폴리텍Ⅱ대학)")

print(f'x값: {x}')  # 경도 출력
print(f'y값: {y}')  # 위도 출력
```

```python
import pandas as pd                          # 엑셀 파일 읽기용 라이브러리
import requests                              # HTTP 요청 라이브러리
from openpyxl import load_workbook          # 기존 엑셀 파일 불러오기
from openpyxl import Workbook               # 새 엑셀 파일 생성
import re                                   # 정규표현식 라이브러리 (괄호 제거용)


filePath = 'c:\data\고등교육기관 하반기 주소록(2025).xlsx'
df_from_excel = pd.read_excel(filePath) 
df_from_excel.columns = df_from_excel.loc[4].tolist() #엑셀의 5번째 줄(인덱스 4)에 실제 컬럼명(학교명, 주소 등)이 있는 경우
df_from_excel = df_from_excel.drop(index=list(range(0,5)))# 5번째 줄(인덱스 4)을 헤더로 바로 인식하고 그 이전 행은 무시, 제목은 이미 만들었으로 인데스4도 제거


url = 'http://api.vworld.kr/req/address'    # VWorld 주소 → 좌표 변환 API 기본 URL
primary_key = '8FBB6D7A-5A0D-3944-819E-F4018E7414F2'  # API 인증키

def request_geo(road, addr_type='ROAD'):    # 주소와 주소유형(기본값: 도로명)을 인자로 받는 함수
    params = {
        'service': 'address',               # 사용할 서비스: 주소 서비스
        'request': 'getcoord',              # 요청 유형: 좌표 가져오기
        'version': '2.0',                   # API 버전
        'crs': 'epsg:4326',                 # 좌표계: WGS84 (위경도 방식)
        'refine': 'true',                   # 불완전한 주소도 자동 보정
        'simple': 'false',                  # 상세 결과 반환
        'format': 'json',                   # 응답 데이터 형식: JSON
        'type': addr_type,                  # 주소 유형: ROAD(도로명) 또는 PARCEL(지번)
        'address': road,                    # 좌표로 변환할 주소값
        'key': primary_key                  # API 인증키
    }
    page = requests.get(url, params=params) # 파라미터를 붙여 GET 요청 전송
    json_data = page.json()                 # 응답 결과를 딕셔너리로 파싱

    if json_data['response']['status'] == 'OK':  # 응답 상태가 정상인 경우
        x = json_data['response']['result']['point']['x']  # 경도 추출
        y = json_data['response']['result']['point']['y']  # 위도 추출
        return x, y                         # 경도, 위도 반환
    else:
        return 0, 0                         # 실패 시 기본값 반환


# 결과 저장용 엑셀 불러오기 (없으면 새로 생성)
try:
    wb = load_workbook(r"c:\data\학교주소좌표.xlsx", data_only=True)  # 기존 파일 열기
    sheet = wb.active                       # 활성화된 시트 선택
except:
    wb = Workbook()                         # 파일 없으면 새 워크북 생성
    sheet = wb.active                       # 기본 시트 선택

university_list = df_from_excel['학교명'].to_list()  # 학교명 컬럼을 리스트로 변환
address_list = df_from_excel['주소'].to_list()       # 주소 컬럼을 리스트로 변환

for num, value in enumerate(address_list):           # 주소 리스트를 인덱스와 함께 순회
    addr = re.sub(r'\([^)]*\)', '', value)           # 괄호와 괄호 안 내용 제거 ex) (구산동) → 삭제
    addr = addr.strip()                              # 앞뒤 공백 제거
    print(addr)                                      # 정제된 주소 출력
    x, y = request_geo(addr)                         # 주소로 좌표 요청
    sheet.append([university_list[num], addr, x, y]) # 학교명, 주소, 경도, 위도를 한 행으로 추가

wb.save(r"c:\data\학교주소좌표.xlsx")               # 결과 엑셀 파일 저장
```

```python
import pandas as pd
import folium

filePath = r'c:\data\학교주소좌표.xlsx'
df_from_excel = pd.read_excel(filePath,engine='openpyxl',header=None)

df_from_excel.columns = ['학교이름','주소','x','y']

name_list = df_from_excel['학교이름'].to_list()
addr_list = df_from_excel['주소'].to_list()
position_x_list = df_from_excel['x'].to_list()
position_y_list = df_from_excel['y'].to_list()

map = folium.Map(location=[37,127],zoom_start=7)

for i in range(len(name_list)):
    if position_x_list[i] != 0:
        marker = folium.Marker([position_y_list[i],position_x_list[i]],
                            popup=name_list[i], 
                            icon = folium.Icon(color='blue'))
        marker.add_to(map) 
map
# map.save(r'c:\data\uni_map.html')
```
