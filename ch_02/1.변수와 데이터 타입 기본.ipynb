```python
original = [1,2,3,[4,5]]
sallow_copy=original[:]
print(original is sallow_copy)
print(original[3] is sallow_copy[3])
```

```python
original = [1,2,3,[4,5]]
sallow_copy=list(original)
print(original[2] is sallow_copy[2])
print(original[1] is sallow_copy[1])
```

```python
original = [1,2,3,[4,5]]
sallow_copy = original.copy()
print(original[2] is sallow_copy[2])
print(original[1] is sallow_copy[1])
```
