### 6장 직접 풀어보기

```python
data['timestamp'] = pd.to_datetime(data['timestamp'])

lunch_data = data[
    (data['timestamp'].dt.hour >= 11) &
    (data['timestamp'].dt.hour <= 14)
]
```

```python
station_sum = (
    lunch_data
    .groupby('station_code')[['people_in', 'people_out']]
    .sum()
)
```

```python
# 필요한 행만 추리기
station_info = station_info[['station.code', 'station.name','geo.latitude','geo.longitude']]
data = data[['station_code','people_in','people_out','timestamp']]

data
```

```python
joined_data = station_sum.join(station_info.set_index('station.code'))
joined_data
```

```python
# 결측제거
joined_data = joined_data.dropna()
```

```python
# 강남, 강북 구분
gangbuk_gangnam = joined_data['region'] = joined_data['geo.latitude'].apply(
    lambda x: '강북' if x >= 37.55 else '강남'
)

gangbuk_gangnam
```

```python
# 승차
top3_in = (
    joined_data
    .sort_values(['region', 'people_in'], ascending=[True, False])
    .groupby('region')
    .head(3)
)
top3_in
```

```python
# 하차
top3_out = (
    joined_data
    .sort_values(['region', 'people_out'], ascending=[True, False])
    .groupby('region')
    .head(3)
)
top3_out
```

```python
# 지도 생성 (서울 중심)
seoul_map = folium.Map(location=[37.55, 126.98], zoom_start=12)

lunch_data_250 = joined_data[joined_data['people_out'] >= 2500000]

for idx, row in lunch_data_250.iterrows():
    folium.Circle(
        location=[row['geo.latitude'], row['geo.longitude']],
        radius=row['people_out'] * 0.0001,  # 0.01%
        color='red',          # 테두리
        fill=True,
        fill_color='red',     # 내부 색
        fill_opacity=0.5
    ).add_to(seoul_map)

seoul_map
```
